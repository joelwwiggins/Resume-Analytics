```python
# load the data
resume = docx2txt.process('ResumeJoelWiggins.docx')
job = docx2txt.process('jobdesc.docx')
```

```python
# printing the resume
resume
```

```python
#print the job description
job
```

```python
# a list of text

text = [resume,job]
```

```python
import pandas
from pandas import DataFrame
df=DataFrame(text)
df.to_clipboard()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
count_matrix = cv.fit_transform(text)
cv.get_feature_names()
```

```python
# import cosine similarity 

from sklearn.metrics.pairwise import cosine_similarity

#print similarity score

print('Similarity score : ',cosine_similarity(count_matrix))
```

```python
# get the percentage
matchpercentage = cosine_similarity(count_matrix)[0][1]
matchpercentage = round(matchpercentage*100,2)
print('Your Resume {} % match to the job description !'.format(matchpercentage))
```
